Como podemos analisar sobre o dataset, percebemos informações contidas porém não são informadas pelo mesmo. Com isso é necessário uma descrição a fim de atenuar o problema sobre as inormações contidas. Deste modo temos:
  
  age: idade em anos;

  sex: 
      gênero

      1 = masculino
      0 =   feminino;
  
  cp: diagnóstico de dores cardíacas (causda por Redução de fluxo sanguíneo)

      1: RFS típica;
      2: RFS atípica;
      3: sem RFS;
      4: assintomático.

  trestbps: pressão arterial;

  fbs: glicose sanguínea;

      1: verdadeiro
      0: falso
    
  restecg: eletrocardiaco;

      0: normal 
      1: normalidade eletrica ST-T
      2: Hipertrofia V. Esquerda
  tchalach: Frequência cardíaca Máx

  exang: RFS por exercício;

      1: sim
      0: não

  oldpeak: depressão de ST induzida por exercício;

  slope: inclinação do segmento ST de pico do exercício:

      1: para cima;
      2: plano;
      3: para baixo.
  
  ca: O número de vasos principais (0-3);
  thal: Uma doença do sangue chamada talassemia: 

      3: normal; 
      6: defeito fixo;
      7: defeito reversível.

  target: doença cardíaca:

      0: não;
      1: sim.

A correlação mostra se os recursos estão relacionados entre si ou com a variável de destino. A correlação pode ser positiva (aumentando um valor, o valor da variável de destino aumenta) ou negativa (aumentando um valor, o valor da variável de destino diminui). A partir deste mapa de calor, podemos ver que a dor no peito "cp" está altamente correlacionada com a variável-alvo. Comparando a relação entre as outras duas variáveis, podemos dizer que a dor no peito é a variável mais útil na previsão da presença de doenças cardíacas. A emergência médica é um ataque cardíaco. O coração geralmente ocorre quando um coágulo sanguíneo impede que o sangue flua para o coração. O tecido perde oxigênio e morre sem sangue, causando dor no peito.

Agora vamos usar os parâmetros da Árvore de Decisão, que são:

entropia: escolher a melhor divisão em cada nó.

```python
#Print tree
def printTree(classifier):
    feature_names = ['Dores Cardíacas', 'Circulação Sanguínea', 
                         'Bloqueio Arterial']
    target_names = ['Sim', 'Não']
    
    #Build the tree
    dot_data = tree.export_graphviz(classifier,                                      
                         out_file=None,feature_names=None,
                         class_names=target_names, filled = True)
    
    #Draw tree
    tr = graphviz.Source(dot_data, format ="png")
    return tr
```

```python
tr_gini = printTree(clf_gini)
tr_gini
```

```python
tr_entropia = printTree(clf_entropy)
tr_entropia
```

Como podemos ver, o "Bloqueio Arterial" é o nó raiz, indicando que "Artérias bloqueadas" é o principal fator que afeta as doenças cardíacas em um paciente.
